```python
glimpse = pd.read_csv('/storage/s97lz6k5/favourita/train.csv', nrows=10)
df_sales = pd.read_csv("/storage/s97lz6k5/favourita/train.csv", usecols=[4], skiprows=range(1,66458909))
```

```python
"""
Explanation:

1) Discard the "id" column. 
2) convert "onpromotion" to boolean variable. 
3) de-skew sales a bit. 
4) Convert "datetime" from string to datetime64 format. 
5) Only starting from 2016-01-01

"""

df_train = pd.read_csv(
    '/storage/s97lz6k5/favourita/train.csv', usecols=[1, 2, 3, 4, 5],
    dtype={'onpromotion': bool},
    converters={'unit_sales': lambda u: np.log1p(
        float(u)) if float(u) > 0 else 0},
    parse_dates=["date"],
    skiprows=range(1, 66458909)  # 2016-01-01
)
```

```python
df_test = pd.read_csv(
    "/storage/s97lz6k5/favourita/test.csv", usecols=[0, 1, 2, 3, 4],
    dtype={'onpromotion': bool},
    parse_dates=["date"]  # , date_parser=parser
).set_index(
    ['store_nbr', 'item_nbr', 'date']
)

items = pd.read_csv(
    "/storage/s97lz6k5/favourita/items.csv",
).set_index("item_nbr")

stores = pd.read_csv(
    "/storage/s97lz6k5/favourita/stores.csv",
).set_index("store_nbr")
```

## Read items and stores

```python
items.head()
```

```python
stores.head()
```

```python
le = LabelEncoder()
items['family'] = le.fit_transform(items['family'].values)

stores['city'] = le.fit_transform(stores['city'].values)
stores['state'] = le.fit_transform(stores['state'].values)
stores['type'] = le.fit_transform(stores['type'].values)
```

```python
items.head()
```

```python
df_2017 = df_train.loc[df_train.date>=pd.datetime(2017,1,1)]
del df_train
```

```python
df_2017.head()
```

## Promo information

```python
promo_2017_train = df_2017.set_index(
    ["store_nbr", "item_nbr", "date"])[["onpromotion"]].unstack(level=-1).fillna(False)
promo_2017_train.columns = promo_2017_train.columns.get_level_values(1)
promo_2017_test = df_test[["onpromotion"]].unstack(level=-1).fillna(False)
promo_2017_test.columns = promo_2017_test.columns.get_level_values(1)
promo_2017_test = promo_2017_test.reindex(promo_2017_train.index).fillna(False)
promo_2017 = pd.concat([promo_2017_train, promo_2017_test], axis=1)
del promo_2017_test, promo_2017_train
```

```python
promo_2017.head()
```

## Convert into time series

```python
df_2017 = df_2017.set_index(
    ["store_nbr", "item_nbr", "date"])[["unit_sales"]].unstack(
        level=-1).fillna(0)
df_2017.columns = df_2017.columns.get_level_values(1)
```
